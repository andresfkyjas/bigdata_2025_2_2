```python
# Notebook: 00_bronze_ingest

from pyspark.sql import functions as F

TARGET_DB = "bakehouse_jobs"

spark.sql(f"CREATE DATABASE IF NOT EXISTS {TARGET_DB}")

# 1) Customers
df_customers = spark.table("samples.bakehouse.sales_customers")
df_customers.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(f"{TARGET_DB}.bronze_customers")

# 2) Franchises
df_franchises = spark.table("samples.bakehouse.sales_franchises")
df_franchises.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(f"{TARGET_DB}.bronze_franchises")

# 3) Suppliers
df_suppliers = spark.table("samples.bakehouse.sales_suppliers")
df_suppliers.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(f"{TARGET_DB}.bronze_suppliers")

# 4) Sales transactions
df_tx = spark.table("samples.bakehouse.sales_transactions")
df_tx.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(f"{TARGET_DB}.bronze_sales_transactions")

# 5) Reviews chunked
df_reviews = spark.table("samples.bakehouse.media_gold_reviews_chunked")
df_reviews.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(f"{TARGET_DB}.bronze_reviews_chunked")
```
